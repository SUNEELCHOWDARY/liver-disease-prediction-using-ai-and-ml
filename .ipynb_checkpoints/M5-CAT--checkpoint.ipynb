# CATBOOSTING CLASSIFIER ALGORITHEM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Data.csv')

df.head()
```

```python
df=df.dropna()
df=df.drop_duplicates()
```

```python
# Transform the columns value(ex: int to str, str to int) for classification purpose.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

var = ['Category','Gender']
  
for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```

```python
df.columns
```

```python
x1 = df.drop(labels='Category', axis=1)
y1 = df.loc[:,'Category']
```

```python
# import imblearn
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

ros =RandomOverSampler(random_state=42)
x,y=ros.fit_resample(x1,y1)
print("OUR DATASET COUNT         : ", Counter(y1))
print("OVER SAMPLING DATA COUNT  : ", Counter(y))
```

```python
!pip install catboost
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42, stratify=y)
print("NUMBER OF TRAIN DATASET    : ", len(x_train))
print("NUMBER OF TEST DATASET      : ", len(x_test))
print("TOTAL NUMBER OF DATASET    : ", len(x_train)+len(x_test))
```

```python
print("NUMBER OF TRAIN DATASET    : ", len(y_train))
print("NUMBER OF TEST DATASET      : ", len(y_test))
print("TOTAL NUMBER OF DATASET    : ", len(y_train)+len(y_test))
```

```python
from catboost import CatBoostClassifier
```

```python
KNN = CatBoostClassifier(
    iterations=50,
    depth=4,
    learning_rate=0.05,
    l2_leaf_reg=50,  # strong regularization
    verbose=0
)
KNN.fit(x_train,y_train)
```

```python
predicted = KNN.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test,predicted)
print('THE CLASSIFICATION REPORT OF RANDOM FOREST CLASSIFIER:\n\n',cr)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predicted)
print('THE CONFUSION MATRIX SCORE OF KNeighborsClassifier:\n\n\n',cm)
```

```python
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(KNN, x, y, scoring='accuracy')
print('THE CROSS VALIDATION TEST RESULT OF ACCURACY :\n\n\n', accuracy*100)
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(y_test,predicted)
print("THE ACCURACY SCORE OF  CATBOOSTING CLASSIFIER IS :",a*100)
```

```python
from sklearn.metrics import hamming_loss
hl = hamming_loss(y_test,predicted)
print("THE HAMMING LOSS OF  CATBOOSTING CLASSIFIER IS :",hl*100)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, title='THE CONFUSION MATRIX SCORE OF CATBOOST CLASSIFIER\n\n', cmap=plt.cm.GnBu):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.xticks(range(cm.shape[1]))
    plt.yticks(range(cm.shape[0]))
    plt.show()

# Compute confusion matrix
cm1 = confusion_matrix(y_test, predicted)

print('THE CONFUSION MATRIX SCORE OF CATBOOST CLASSIFIER:\n\n')
print(cm1)

# Plot confusion matrix with updated color map
plot_confusion_matrix(cm1)
```

```python
import matplotlib.pyplot as plt
df2 = pd.DataFrame()
df2["y_test"] = y_test
df2["predicted"] = predicted
df2.reset_index(inplace=True)
plt.figure(figsize=(20, 5))
plt.plot(df2["predicted"][:100], marker='x', linestyle='dashed', color='red')
plt.plot(df2["y_test"][:100],  marker='o', linestyle='dashed', color='green')
plt.show()
```
